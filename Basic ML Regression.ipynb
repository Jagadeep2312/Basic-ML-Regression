```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv("Gym_Diet.csv")
print("‚úÖ Dataset Loaded Successfully")

food_df = df[["Protein_g", "Carbs_g", "Fats_g", "Food_Calories"]].dropna()

X_food = food_df[["Protein_g", "Carbs_g", "Fats_g"]]
y_food = food_df["Food_Calories"]

food_model = LinearRegression()
food_model.fit(X_food, y_food)

workout_df = df[["Workout_Duration_Min", "Calories_Burned"]].dropna()

X_workout = workout_df[["Workout_Duration_Min"]]
y_workout = workout_df["Calories_Burned"]

workout_model = LinearRegression()
workout_model.fit(X_workout, y_workout)

while True:
    print("\nüèãÔ∏è GYM & DIET ANALYSIS SYSTEM")
    print("1. Predict Food Calories")
    print("2. Predict Workout Calories Burned")
    print("3. BMI + Daily Calories + Workout Plan")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        protein = float(input("Enter protein (grams): "))
        carbs = float(input("Enter carbs (grams): "))
        fats = float(input("Enter fats (grams): "))

        calories = food_model.predict([[protein, carbs, fats]])
        print(f"\nüçΩÔ∏è Predicted Food Calories: {calories[0]:.2f} kcal")

    elif choice == "2":
        duration = float(input("Enter workout duration (minutes): "))
        burned = workout_model.predict([[duration]])

        print(f"\nüî• Estimated Calories Burned: {burned[0]:.2f} kcal")

    elif choice == "3":
        weight = float(input("Enter weight (kg): "))
        height = float(input("Enter height (meters): "))
        age = int(input("Enter age: "))
        gender = input("Enter gender (male/female): ").lower()
        activity = input("Activity level (low/medium/high): ").lower()

        bmi = weight / (height ** 2)

        if bmi < 18.5:
            bmi_status = "Underweight"
            workout_level = "Beginner"
        elif bmi < 25:
            bmi_status = "Normal Weight"
            workout_level = "Intermediate"
        elif bmi < 30:
            bmi_status = "Overweight"
            workout_level = "Advanced"
        else:
            bmi_status = "Obese"
            workout_level = "Advanced"

        if gender == "male":
            bmr = 10 * weight + 6.25 * (height * 100) - 5 * age + 5
        else:
            bmr = 10 * weight + 6.25 * (height * 100) - 5 * age - 161

        if activity == "low":
            daily_calories = bmr * 1.2
        elif activity == "medium":
            daily_calories = bmr * 1.55
        else:
            daily_calories = bmr * 1.9

        print("\nüìä HEALTH REPORT")
        print(f"üëâ BMI Value           : {bmi:.2f}")
        print(f"üëâ BMI Category        : {bmi_status}")
        print(f"üëâ Daily Calories Need : {daily_calories:.0f} kcal")

        print("\nüèãÔ∏è Recommended Workout Plan:")
        if workout_level == "Beginner":
            print("- Walking / Cycling (30 min)")
            print("- Light bodyweight exercises")
            print("- 3 days per week")
        elif workout_level == "Intermediate":
            print("- Strength training")
            print("- Cardio 20‚Äì30 min")
            print("- 4‚Äì5 days per week")
        else:
            print("- HIIT workouts")
            print("- Heavy strength training")
            print("- 5‚Äì6 days per week")

    elif choice == "4":
        print("üëã Thank you for using the system")
        break

    else:
        print("‚ùå Invalid choice, try again")
```
